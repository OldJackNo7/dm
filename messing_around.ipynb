```python
import pandas as pd
import numpy as np

app_id = 1151640
data_path = 'data/'

positive_reviews = pd.read_csv(data_path + 'review_%d_positive.csv' % app_id)
negative_reviews = pd.read_csv(data_path + 'review_%d_negative.csv' % app_id)
summary = pd.read_csv(data_path + 'summary_%d.csv' % app_id)
```

```python
from enum import Enum
class Sentiment(Enum):
    POSITIVE = 'positive'
    NEGATIVE = 'negative'
    NEUTRAL = 'neutral'

features = ['votes_up', 'votes_funny', 'weighted_vote_score', 'author_num_games_owned', 'author_playtime_at_review', 'label']

# Label our datasets
positive_reviews['label'] = Sentiment.POSITIVE
negative_reviews['label'] = Sentiment.NEGATIVE

# Create the batch of training data
training_data = positive_reviews.append(negative_reviews)
```

```python
def euclidean_distance(sx, sy):
    return np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(sx, sy)]))

no_rows = features.shape[0]
distances = [[0 for _ in range(no_rows)] for _ in range(no_rows)]

for x in range(no_rows):
    for y in range(no_rows):
        if y == x:
            continue
        distances[x][y] = euclidean_distance(features.loc[x], features.loc[y])
        
distances
```
